```python
#load the relevant packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
#access the dataset
insurance=pd.read_csv("insurance.csv")
insurance.head()
```

```python
insurance.shape
```

```python
#check data type
insurance.dtypes
```

```python
#how age affects insurance costs
# The iloc function allows use to select rows from the dataframe.  
x = insurance.iloc[:,:1].values
y = insurance.iloc[:,-1].values
plt.figure(figsize=(15,10), dpi = 100)
plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
#defining the model & fit model
charges_model=LinearRegression()
charges_model.fit(x,y)
```

```python

plt.figure(figsize=(15,10), dpi = 100)
y_pred = charges_model.predict(x)
plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.ylabel('Age')
plt.xlabel('Charges')
plt.show()

print("""Of the variables in the dataset age has the strongest relationship  of all even though its fairly weak 
as show on the plot above""")
```

```python
# The iloc function allows use to select rows from the dataframe.  
x=insurance.loc[:,'bmi'].values
y=insurance.loc[:,'charges'].values
plt.figure(figsize=(15,10), dpi = 100)
plt.scatter(x,y ,color = 'b')
plt.xlabel('BMI')
plt.ylabel('CHARGES')
plt.show()
```

```python
#reshape the array
x=x.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
#defining the model & fit model
bmi_charges=LinearRegression()
bmi_charges.fit(x,y)
```

```python
plt.figure(figsize=(15,10), dpi = 100)
y_pred = bmi_charges.predict(x)
plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.ylabel('BMI')
plt.xlabel('Charges')
plt.show()
print('BMI has a weak realationship with the insurance costs as shown on the plot')
```
